```python
import cv2 
from matplotlib import pyplot as plt
import numpy as np
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
print(face_cascade)
```

```python
image = cv2.imread('class-pic.jpg')
```

```python
# display the original pic
image_show = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image_show)
plt.show()
```

```python
# conversion to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = np.array(gray, dtype='uint8')
```

```python
# detect faces is grayScale image
faces = face_cascade.detectMultiScale(gray, 1.1, 5)
```

```python
for (x, y, w, h) in faces:
   cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
   roi_gray = gray[y:y+h, x:x+w]
   roi_color = image[y:y+h, x:x+w]
   plt.imshow(roi_gray)
   plt.show()
```

```python
# detect faces is grayScale image
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = image[y:y+h, x:x+w]
    plt.imshow(roi_gray)
    plt.show()
```

The detectMultiScale function in OpenCV is used for object detection, including face detection. It detects objects of different sizes in the input image and returns a list of rectangles that represent the detected objects. Here is an explanation of the common parameters used in detectMultiScale for face detection:

scaleFactor: It specifies the factor by which the image size is reduced at each image scale. For example, a scaleFactor of 1.1 means the image is reduced by 10% at each scale. A smaller scaleFactor will increase the detection time but may increase the chances of detecting smaller faces.

minNeighbors: It specifies the minimum number of neighbors (rectangles) that an object candidate should have to be retained. This parameter helps filter out false detections. Increasing the value will make the detection more selective but may miss some valid detections.

minSize: It specifies the minimum object size to be detected. It is usually specified as a tuple (width, height) in pixels. Objects smaller than this size will be ignored. It helps speed up the detection process and eliminates detecting very small false positives.

maxSize: It specifies the maximum object size to be detected. Similar to minSize, it is also specified as a tuple (width, height) in pixels. Objects larger than this size will be ignored. It helps speed up the detection process and eliminates detecting very large false positives.

flags: It is an optional parameter that can be used to modify the behavior of the detection algorithm. Some common flags include cv2.CASCADE_SCALE_IMAGE, which scales the image instead of the detector for better performance, and cv2.CASCADE_DO_ROUGH_SEARCH, which performs a faster but less accurate initial scan.

These parameters can be adjusted to optimize the face detection process based on the specific requirements of the application, the characteristics of the input images, and the desired trade-off between accuracy and speed.
