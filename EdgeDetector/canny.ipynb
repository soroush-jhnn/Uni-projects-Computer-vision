```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load an image
img = cv2.imread('images/camera_256.jpg', 0)
```

```python
# display original image
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# Apply Roberts operator
roberts_op = cv2.filter2D(img, -1, np.array([[1, 0], [0, -1]]))
```

```python
# display image with Roberts operator
plt.imshow(roberts_op, cmap='gray')
plt.show()
```

```python
# Apply Prewitt operator
prewitt_x = cv2.filter2D(img, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
prewitt_y = cv2.filter2D(img, -1, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))
prewitt_op = np.sqrt(prewitt_x**2 + prewitt_y**2)
```

```python
# display image with Prewitt operator
plt.imshow(prewitt_op, cmap='gray')
plt.show()
```

```python
# Apply Sobel operator
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel_op = np.sqrt(sobel_x**2 + sobel_y**2)
```

```python
# display image with Sobel operator
plt.imshow(sobel_op, cmap='gray')
plt.show()
```

```python
# Apply Canny operator
canny_op = cv2.Canny(img, 100, 200)
```

```python
# display image with Canny operator
plt.imshow(canny_op, cmap='gray')
plt.show()
```
