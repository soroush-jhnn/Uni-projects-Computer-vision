```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Open a simple image from Image folder
img = cv2.imread("Images/Cup.JPG")
```

```python
# demonstrate the original pic
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image_RGB)
plt.show()
```

```python
# shirnking original image with Nearest Neighborhood method
shrinkNN = cv2.resize(
    img, (int(img.shape[1]*0.3), int(img.shape[0]*0.3)), interpolation=cv2.INTER_NEAREST)
```

```python
# magnifying original image with Nearest Neighborhood method
magnifyNN = cv2.resize(
    img, (int(img.shape[1]*2.3), int(img.shape[0]*2.3)), interpolation=cv2.INTER_NEAREST)
```

```python
# show the converted images
plt.imshow(cv2.cvtColor(shrinkNN, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv2.cvtColor(magnifyNN, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# saving the converted images
cv2.imwrite("Images/0.3_cup_nn.jpg", shrinkNN)
cv2.imwrite("Images/2.3_cup_nn.jpg", magnifyNN)
```

```python
# shirnking original image with Bilinear method
shrinkBi = cv2.resize(
    img, (int(img.shape[1]*0.3), int(img.shape[0]*0.3)), interpolation=cv2.INTER_LINEAR)
```

```python
# magnifying original image with Bilinear method
magnifyBi = cv2.resize(
    img, (int(img.shape[1]*2.3), int(img.shape[0]*2.3)), interpolation=cv2.INTER_LINEAR)
```

```python
# show the converted images
plt.imshow(cv2.cvtColor(shrinkBi, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv2.cvtColor(magnifyBi, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# saving the converted images
cv2.imwrite("Images/0.3_cup_Bi.jpg", shrinkBi)
cv2.imwrite("Images/2.3_cup_Bi.jpg", magnifyBi)
```

```python
# shirnking original image with Bicubic method
shrinkBc = cv2.resize(
    img, (int(img.shape[1]*0.3), int(img.shape[0]*0.3)), interpolation=cv2.INTER_CUBIC)
```

```python
# magnifying original image with Bicubic method
magnifyBc = cv2.resize(
    img, (int(img.shape[1]*2.3), int(img.shape[0]*2.3)), interpolation=cv2.INTER_CUBIC)
```

```python
# show the converted images
plt.imshow(cv2.cvtColor(shrinkBc, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv2.cvtColor(magnifyBc, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# saving the converted images
cv2.imwrite("Images/0.3_cup_Bc.jpg", shrinkBc)
cv2.imwrite("Images/2.3_cup_Bc.jpg", magnifyBc)
```
